```python
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input,Embedding, SimpleRNN,LSTM, Dense, Dropout
```

```python
sentences = [
 "I love this product",
 "This movie made me smile",
 "Service was friendly and quick",
 "Today felt bright and happy",
 "This is the best day",
 "Absolutely fantastic experience",
 "I enjoyed every single moment",
 "Great job, well done",
 "The food tasted delicious",
 "Totally recommend to everyone",
 "Very satisfied with results",
 "This worked better than expected",
 "Amazing quality and value",
 "Such a pleasant surprise",
 "I feel positive about this",
 "The weather is lovely",
 "This song makes me happy",
 "A wonderful start to the day",
 "Highly impressed with the quality",
 "Everything went perfectly",
 "I hate this product",
 "This movie bored me",
 "Service was rude and slow",
 "Today was cold and lonely",
 "This is the worst day",
 "Terrible experience overall",
 "I regret buying this",
 "Very disappointed with results",
 "The food tasted awful",
 "Do not recommend this",
 "It broke after one use",
 "Not worth the money",
 "Utterly frustrating and annoying",
 "I feel negative about this",
 "Such a waste of time",
 "The service was terrible",
 "This book is so boring",
 "What a dreadful day",
 "I'm very displeased with this",
 "Couldn't be worse"
]
labels = [1]*20 + [0]*20
labels = np.array(labels)
print(labels)
```

```python
vocab_size = 2000
tok = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tok.fit_on_texts(sentences)
word_index = tok.word_index
seq = tok.texts_to_sequences(sentences)
# print(seq)
X = pad_sequences(seq, maxlen=10, padding="post", truncating="post")
# print(X)
```

```python
embed_dim = 16
rnn_units = 8
```

```python
inp = Input(shape= (10,),dtype="int32",name = 'input')
x = Embedding(input_dim=vocab_size,output_dim=embed_dim,mask_zero=True,name = 'embed')(inp)
rnn = SimpleRNN(units = rnn_units, return_sequences = False,return_state= False,name = 'simple_rnn')
x_last = rnn(x)
out = Dense(1,activation='sigmoid',name = 'out')(x_last)
model = Model(inputs =inp,outputs = out)
model.compile(optimizer='adam',loss = 'binary_crossentropy',metrics=['accuracy'])
model.summary()

```

```python
from tensorflow.keras.models import Sequential

model_seq = Sequential([
    Input(shape=(10,), dtype="int32", name='input'),
    Embedding(input_dim=vocab_size, output_dim=embed_dim, mask_zero=True, name='embed'),
    SimpleRNN(units=rnn_units, return_sequences=False, name='simple_rnn'),
    Dense(1, activation='sigmoid', name='out')
])

model_seq.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model_seq.summary()
```

```python
def predict_sentiment(text):
  seq = tok.texts_to_sequences([text])
  padded_seq = pad_sequences(seq, maxlen=10, padding='post')

  prediction = model.predict(padded_seq)[0][0]

  if prediction > 0.5:
    return f"Positive (Confidence: {prediction:.2f})"
  else:
    return f"Negative (Confidence: {1 - prediction:.2f})"

user_sentence = input("Enter a sentence to predict its sentiment: ")

sentiment = predict_sentiment(user_sentence)
print(f"The sentiment of \'{user_sentence}\' is: {sentiment}")
```
